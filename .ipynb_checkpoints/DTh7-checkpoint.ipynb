```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
percentage = 0.05 # Годовой процент по кредиту
```

```python
mean_income, std_income = 50000, 30000
people_count = 200
income = np.abs(np.random.normal(mean_income, std_income, people_count))

credits = np.array([inc * np.random.randint(3, 6) + np.random.rand() * inc for inc in income])
```

```python
df = pd.DataFrame(columns=['index', 'income', 'credit'])
```

```python
df['index'] = np.arange(0, people_count)
df['income'] = np.round(income)
df['credit'] = np.round(credits)
```

```python
plt.scatter(income, credits)
plt.show()
```

$$RSS=\sum_{i}(y_{i} - (a + bx_{i})^{2})$$

$$\begin{cases}
b \sum_{i} x_{i}^{2} + a \sum_{i} x_{i} = \sum_{i}x_{i} y_{i} \\
b \sum_{i} x_{i} + na = \sum_{i} y_{i}
\end{cases}$$

<p>
Будем рассматривать отклонение по y (т.е. заявленная сумма кредита)

```python
sum_income2 = np.sum(np.square(income))
sum_income = np.sum(income)
credits_income = np.sum(credits * income)
sum_credits = np.sum(credits)
```

```python
matrix = np.array([[sum_income2, sum_income], 
                   [sum_income, people_count]])
free_values = np.array([credits_income, sum_credits])
```

```python
result = np.linalg.solve(matrix, free_values)
```

```python
x = np.linspace(0, max(income), 100)
y = lambda x: result[0] * x + result[1]
```

```python
plt.scatter(income, credits)
plt.plot(x, y(x), color='red', linewidth=2)
plt.show()
```

```python
a, b = result[1], result[0]
y_mean = credits.mean()
y_hat = np.array([a + b * inc for inc in income])

rss = np.sum((credits - y_hat) ** 2)
tss = np.sum((credits - y_mean) ** 2)
ess = np.sum((y_hat - y_mean) ** 2)

determination = ess / tss
korrel_pirson = (np.sum((credits - y_mean) * (y_hat - y_mean)) / 
                 np.sqrt(tss * ess))
```

```python
print(rss, tss, ess)
```

```python
print(determination, korrel_pirson)
```
